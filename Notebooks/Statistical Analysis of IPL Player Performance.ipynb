```python
def calculate_adjusted_scores(df):
  """
  Calculates adjusted scores for "not-out" innings in an IPL ball-by-ball dataset.

  Args:
      df (pd.DataFrame): IPL ball-by-ball data with columns for 'batter',
                          'runs_scored', and 'is_out' (or equivalent dismissal indicator).

  Returns:
      pd.DataFrame: Modified DataFrame with a new column 'adjusted_score'.
  """

  adjusted_df = df.copy()

  # Group by batter and create a window for calculating cumulative runs
  for player, player_df in adjusted_df.groupby('batter'):
    player_df['cum_runs'] = player_df['batsman_run'].cumsum()
    dismissed = player_df['isWicketDelivery']  # Assuming 'is_out' indicates dismissal

    for index, row in player_df.iterrows():
      runs_scored = row['cum_runs']
      is_out = row['isWicketDelivery']

      if not is_out:  # Handle "not-out" situations
        # Look back window for previous innings (excluding current delivery)
        look_back = player_df.loc[:index, 'isWicketDelivery']

        # Calculate number of previous innings with higher scores (ni_j)
        ni_j = (look_back == True).sum()

        # Calculate conditional average for non-dismissed cases (Ci_k)
        ci_j = 0 if ni_j == 0 else player_df.loc[:index, 'cum_runs'][look_back == True].sum()

        # Adjusted score using conditional average (Ei_j)
        adjusted_score = ci_j / ni_j if ni_j > 0 else player_df['cum_runs'].iloc[index]
        adjusted_df.at[index, 'adjusted_score'] = adjusted_score

  return adjusted_df
```

```python
import pandas as pd

def transform_dataset(df):
    # Group by match ID and batsman
    grouped_df = df.groupby(['ID', 'batter'])
    
    # Calculate total runs per match for each batsman
    batsman_runs = grouped_df['batsman_run'].sum().reset_index(name='runs')
    
    # Check if the batsman was dismissed in each match
    dismissal_status = grouped_df['isWicketDelivery'].any().reset_index(name='dismissed')
    
    # Merge runs and dismissal status
    transformed_df = pd.merge(batsman_runs, dismissal_status, on=['ID', 'batter'], how='left')
    
    return transformed_df

# Apply the transformation function to your dataset
df_transformed = transform_dataset(ipl_ball_by_ball.copy())  # Operate on a copy to avoid modifying original data

# Print or use the transformed DataFrame (df_transformed) for further analysis
print(df_transformed.head())
```

```python
def calculate_cumulative_probabilities(adjusted_scores):
  """
  Calculates the cumulative probability of exceeding each score in a list.

  Args:
      adjusted_scores (list): List of adjusted scores.

  Returns:
      list, list: Sorted adjusted scores and corresponding cumulative probabilities.
  """

  sorted_scores = sorted(adjusted_scores)
  cumulative_probs = []
  for i, score in enumerate(sorted_scores):
      count_greater = len([s for s in adjusted_scores if s > score])
      cumulative_prob = (count_greater + 1) / (len(adjusted_scores) + 1)  # Add 1 for smoothing
      cumulative_probs.append(cumulative_prob)

  return sorted_scores, cumulative_probs
```

```python
# Assuming 'adjusted_scores_dhoni' and 'adjusted_scores_kohli' are columns in your DataFrame containing the adjusted scores for Dhoni and Kohli, respectively
sorted_scores_dhoni, cumulative_probs_dhoni = calculate_cumulative_probabilities(adjusted_scores_dhoni)
sorted_scores_kohli, cumulative_probs_kohli = calculate_cumulative_probabilities(adjusted_scores_kohli)

plt.plot(sorted_scores_dhoni, cumulative_probs_dhoni, label='MS Dhoni')
plt.plot(sorted_scores_kohli, cumulative_probs_kohli, label='Virat Kohli')
plt.xlabel('Adjusted Scores')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Probability of Adjusted Scores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import hmean

def calculate_cbr(runs_conceded, wickets_taken, balls_bowled):
    """
    Calculates the Combined Bowling Rate (CBR) score for a bowler.

    Args:
        runs_conceded (int): Total number of runs conceded by the bowler.
        wickets_taken (int): Total number of wickets taken by the bowler.
        balls_bowled (int): Total number of balls bowled by the bowler.

    Returns:
        float: The calculated CBR score.
    """

    if wickets_taken == 0:  # Handle case where no wickets are taken
        return 0  # No wickets taken, CBR is 0

    # Calculate bowling average
    bowling_average = runs_conceded / wickets_taken

    # Calculate economy rate, handle division by zero by setting it to a large value
    economy_rate = runs_conceded / (balls_bowled / 6) if balls_bowled > 0 else float('inf')

    # Calculate adjusted strike rate, handle division by zero by setting it to a large value
    adjusted_strike_rate = (runs_conceded * wickets_taken) / balls_bowled if balls_bowled > 0 else float('inf')

    # Calculate CBR using the harmonic mean of bowling average, economy rate, and adjusted strike rate
    cbr = hmean([bowling_average, economy_rate, adjusted_strike_rate])

    return cbr

# Test the function
print(calculate_cbr(100, 5, 0))  # Example usage
```

```python
def calculate_cbr(runs_conceded, wickets_taken, balls_bowled):
    if wickets_taken == 0 or balls_bowled == 0:
        return 100

    bowling_average = runs_conceded / wickets_taken
    economy_rate = runs_conceded / (balls_bowled / 6)
    adjusted_strike_rate = balls_bowled / wickets_taken
    
    if economy_rate == 0 or adjusted_strike_rate == 0:
        return 100

    cbr = 3 / (1 / bowling_average + 1 / economy_rate + 1 / adjusted_strike_rate)
    return cbr
```
